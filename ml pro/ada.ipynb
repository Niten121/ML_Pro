```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeRegressor
```

```python
dataset = pd.read_csv("/content/drive/MyDrive/Copy of car data.csv")
```

```python
dataset.head()
```

```python
l=LabelEncoder()
```

```python
label=l.fit_transform(dataset.Fuel_Type)
```

```python
label1=l.fit_transform(dataset.Car_Name)
```

```python
dataset["outcome"]=label
```

```python
dataset["car_type"]=label1
```

```python
dataset.head()
```

```python
x= dataset.drop(['Selling_Price','Transmission','Seller_Type','Car_Name','Fuel_Type'], axis='columns')
y = dataset.Selling_Price
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 100)
```

```python
from sklearn.ensemble import AdaBoostRegressor
```

```python
base_reg = DecisionTreeRegressor(max_depth=3)
ada_reg = AdaBoostRegressor(base_estimator = base_reg,n_estimators = 50,learning_rate = 0.2,random_state = 1)
```

```python
ada_reg.fit(x_train,y_train)
y_pred = ada_reg.predict(x_test)
y_pred
```

```python
score = ada_reg.score(x_test,y_test)
score
```
